```python
class Node():
    def __init__(self,value):
        self.value = value
        self.next = None
        

        
class Linkedlist():
    
    def __init__(self):
        self.head = None
        

    def addFront(self,value):
        new = Node(value)
        if self.head is None:
            self.head = new
            return
        else:
            new.next =  self.head
            self.head = new
            
        
    def addEnd(self,value):
        new = Node(value)
        this = self.head
        if self.head is None:
            return print("empty list")
        while this.next:
            this = this.next
        this.next = new
            
            
    def delete(self, d):
        this = self.head
        
        if this and this.value == d:
            self.head = this.next
            this = None
            return
        prev = None
        while this and this.value != d:
            prev = this
            this = this.next  
        if this is None:
            return print(d,"is Not in list")
        else:
            prev.next = this.next
            this = None
            
            
    def delete_pos(self, pos):
        this = self.head
        count = 1
        if pos == 1:
            self.head = this.next
            this = None
            return
        prev = None
        while this and count != pos:
            prev = this
            this = this.next
            count += 1
        if this is None:
            return print(pos,"Limit exceeds")
        else:
            prev.next = this.next
            this = None
            
        
    def Print(self):
        this = self.head
        while this:
            print(this.value, end = "->")
            this = this.next
        print("None")
        print()
        
        
    def __len__(self):
        count = 0
        this = self.head
        while this:
            count += 1
            this = this.next
        return count
        
        
        
m = Linkedlist()
m.addFront(1)
m.addFront(2)

m.addEnd(5)

m.addFront(9)

m.addEnd(7)

m.Print()

m.delete(9)

m.Print()

m.delete(5)

m.Print()

m.delete(7)

m.Print()

m.addFront(21)
m.addFront(15)
m.addFront(10)
m.addFront(13)

m.addEnd(8)
m.addEnd(20)
m.addEnd(41)

m.Print()

m.delete_pos(20)

m.Print()

m.delete_pos(1)

m.Print()

m.delete_pos(3)

m.Print()

len(m)
```
